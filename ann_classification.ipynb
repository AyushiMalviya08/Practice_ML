```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.read_csv("/content/BankMarketing.csv", sep=";")
```

```python
x=df.drop('y',axis=1)
Y=df['y']
```

```python
one_hot_encoded_X= pd.get_dummies(x, columns=['job','marital','education','default','housing','loan','contact','month','poutcome'], dtype=int)
```

```python
one_hot_encoded_X
```

```python
one_hot_encoded_Y=pd.get_dummies(Y,dtype=int)
```

```python
one_hot_encoded_Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(one_hot_encoded_X,one_hot_encoded_Y, test_size =0.3)
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
```

```python
scaledX_train = scale.fit_transform(x_train)
```

```python
scaledX_test = scale.fit_transform(x_test)
```

```python
scaledy_train = scale.fit_transform(y_train)
```

```python
scaledy_test = scale.fit_transform(x_test)
```

```python
scaledX_train
```

```python
scaledX_test
```

```python
scaledy_test
```

```python
scaledy_train
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense

```

```python
model = Sequential()
model.add(Dense(200,input_dim=scaledX_train.shape[1], activation='relu'))
model.add(Dense(200, activation='relu'))
model.add(Dense(2, activation='sigmoid'))


```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(scaledX_train,scaledy_train,batch_size=32,epoch=10)
```

```python
y_pred=model.predict(scaledX_train)
```
